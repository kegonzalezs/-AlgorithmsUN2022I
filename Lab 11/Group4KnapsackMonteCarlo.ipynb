##Funciones y variables a usar

```python
import numpy as np
import itertools

#Chequeo que el conjunto de items no sobrepase la capacidad de la mochila
def checkWeight(x,a,b):
  return np.sum(np.multiply(x, a)) <= b

#Solucin exacta para el punto 1
def solve(a,b):
  k=0
  comb = list(itertools.product([0, 1], repeat=a.size))
  for i in list(comb):
    if (checkWeight(a,i,b)):
      k+=1
  print("Para el vector",a)
  print("Con una mochila de capacidad ",b,"\n" )
  print("La cantidad de soluciones exactas es ",k)
  print("La cantidad de combinaciones es ",2**len(a))
  print("La proporcin de es ",k/2**len(a))

#Solucin monte carlo para puntos 2 y 3

def solveMonteCarlo(nRand,a,b):
  k=0
  for i in nRand:
    if checkWeight(i,a,b):
      k+=1
  print("Para el vector",a)
  print("Con una mochila de capacidad ",b,"y usando MonteCarlo con ",len(nRand)," nmeros aleatorios\n" )
  print("La cantidad de soluciones obtenidas es ",k)
  print("La cantidad de combinaciones es ",len(nRand))
  print("La proporcin de es ",k/len(nRand),"\n")


#Entradas para el punto 2

#lista de vectores binarios aleatorios de 1000 y 10000 de tamao 10

ve1000=[[np.random.choice([0,1]) for i in range(10)] for i in range(1000)]

ve10000=[[np.random.choice([0,1]) for i in range(10)] for i in range(10000)]

a10=np.array(range(1,10+1))




#Entradas para el punto 3

#lista de vectores binarios aleatorios de 10000, 100000 y 1000000 de tamao 50

v10000=[[np.random.choice([0,1]) for i in range(50)] for i in range(10000)]

v100000=[[np.random.choice([0,1]) for i in range(50)] for i in range(100000)]

v1000000=[[np.random.choice([0,1]) for i in range(50)] for i in range(100000)]

a=np.array(range(1,50+1))


```

##1. Count and calculate the exact proportion of  Knapsack solutions. for the problem in the image.

```python
solve(np.array([1, 1, 2, 4, 12]),15)
```

##2. Estimate the proportion and number of Knapsack solutions. if the sizes are (1,2,3,4,5,6,7,9,10) and the capacity of the knapsack is 10 using Mote Carlo with 1.000, 10.000 random binary vectors.

```python
solveMonteCarlo(ve1000,a10,10)
solveMonteCarlo(ve10000,a10,10)
solve(a10,10)
```

##3. Estimate the proportion and number of Knapsack solutions. if the sizes are (1,2,3,4,, 49,50) and the capacity of the knapsack are 10, 50, 100, 1275 using Mote Carlo with 10.000, 100.000 and 1.000.000 random binary vectors.

```python
vec=[v10000,v100000,v1000000]
capacidad=[10,50,100,1275]

for i in capacidad:
  print("**********************************************************************************\n")
  print("Con capacidad ",i,"\n")
  for j in vec:
    solveMonteCarlo(j,a,i)
  
```
